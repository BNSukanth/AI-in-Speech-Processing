```python
import librosa

y, sr = librosa.load("SpeechRec.wav") #Loading the audio 

print(y)
print(sr)
```

```python
librosa.display.waveshow(y,color = 'blue') #plotting the audio
```

```python
from IPython.display import Audio #playing the Audio
Audio(data=y, rate=sr)
```

```python
Audio(y[11000:15000],rate = sr) # the sound segment of the vowel 'a'
```

```python
import numpy as np
import matplotlib.pyplot as plt

vowel_a = y[11000:15000]
vowel_a_fft = np.fft.fft(vowel_a) # fft on the sound segment of the vowel 'a'

plt.figure(figsize=(20,5))#plotting
plt.plot(np.abs(vowel_a_fft))
plt.title('FFT of the vowel A')

```

```python
def generate_spectrogram(signal, sample_rate):
    plt.specgram(signal, Fs=sample_rate)
    plt.title("Spectrogram of Speech Signal") #plotting spectogram
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")
    plt.colorbar()

# Generate spectrogram
plt.figure(figsize=(12, 6))
generate_spectrogram(y,sr)
plt.show()
```

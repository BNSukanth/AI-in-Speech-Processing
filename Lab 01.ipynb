```python
import librosa

y, sr = librosa.load("SpeechRec.wav")
```

```python
librosa.display.waveshow(y,color = 'blue')
```

```python
print(y)
print(sr)
# for spoken words we can see the change in the amplitude of the audio plotted, 
# and for no words spken there is relatively less change in the amplitude.
```

```python
from IPython.display import Audio
```

```python
Audio(data=y, rate=sr)
```

```python
import numpy as np
reverse_speech=np.flipud(y)
display(Audio(data=reverse_speech,rate=sr))
```

```python
librosa.display.waveshow(reverse_speech,color = 'blue')
```
